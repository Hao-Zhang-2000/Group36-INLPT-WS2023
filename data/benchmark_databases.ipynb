```python
import json
from databases import PineconeDataBase, NumpyDataBase
from embedding_models import EmbeddingModelMiniLML6
from time import perf_counter
```

```python
with open("data.json", "r") as f:
    data = json.load(f)
```

```python
embedding_model = EmbeddingModelMiniLML6()
pinecone_database = PineconeDataBase(embedding_model=embedding_model)
numpy_database = NumpyDataBase(data=data, embedding_model=embedding_model)
```

```python
queries = [
    "Are there Artificial Neural Networks that can detect or predict cancer?",
] * 100
```

```python
lengths_numpy = {}
lengths_pinecone = {}

for k in range(1, 11):
    lengths_numpy[k] = []
    lengths_pinecone[k] = []
    
    for query in queries:
        # we specifically do the calls one after another so that numpy will not benefit from warmed up cache
        start = perf_counter()
        results = numpy_database.retrieve_by_query(query, top=k)
        end = perf_counter()
        lengths_numpy[k].append(end-start)
        
        start = perf_counter()
        results = pinecone_database.retrieve_by_query(query, top=k)
        end = perf_counter()
        
        lengths_pinecone[k].append(end-start)
        
```

```python
from matplotlib import pyplot as plt
from copy import deepcopy
import numpy as np
```

```python

dict1 = deepcopy(lengths_numpy)
dict2 = deepcopy(lengths_pinecone)

# Function to calculate max, mean, min
def get_stats(dictionary):
    stats = {key: [np.max(dictionary[key]), np.mean(dictionary[key]), np.min(dictionary[key])] for key in dictionary}
    return stats

# Get statistics for both dictionaries
stats1 = get_stats(dict1)
stats2 = get_stats(dict2)

# Preparing data for plotting
keys = list(stats1.keys())
stats1_values = np.array(list(stats1.values()))
stats2_values = np.array(list(stats2.values()))

# Max, mean, min for each key in both dictionaries
max1, mean1, min1 = stats1_values[:,0], stats1_values[:,1], stats1_values[:,2]
max2, mean2, min2 = stats2_values[:,0], stats2_values[:,1], stats2_values[:,2]

# Plotting
x = np.arange(len(keys))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(x - width/2, max1, width, label='Max Dict1', alpha=0.8)
rects2 = ax.bar(x + width/2, max2, width, label='Max Dict2', alpha=0.8)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Values')
ax.set_title('Max values by key and dictionary')
ax.set_xticks(x)
ax.set_xticklabels(keys)
ax.legend()

# Function to auto-label the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height,2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()

# Showing the plot for max values
plt.show()
```
